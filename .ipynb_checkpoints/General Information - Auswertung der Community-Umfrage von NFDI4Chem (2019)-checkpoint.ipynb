```python
# Definiere die NFDI4Chem-Farben
petrol_rgb = (0/255, 156/255, 188/255)
orange_rgb = (238/255, 116/255, 0/255)
yellow_rgb = (241/255, 222/255, 30/255)

# Definiere die RGBA-Werte für die Farbe "petrol" in verschiedenen Transparenzstufen
petrol_rgba_100 = (*petrol_rgb, 1.0)  # 100% Transparenz
petrol_rgba_75 = (*petrol_rgb, 0.75)  # 75% Transparenz
petrol_rgba_50 = (*petrol_rgb, 0.50)  # 50% Transparenz
petrol_rgba_25 = (*petrol_rgb, 0.25)  # 25% Transparenz

NFDI4Chem = ('#009CBC', '#EE7400', '#F1DE1E', '#7f7f7f')
NFDI4Chem_trans = ('#009CBC', '#EE7400', '#F1DE1E','#4c4c4c', '#009CBCB2', '#EE7400B2', '#F1DE1EB2', '#7f7f7f', '#009CBC4D', '#EE74004D', '#F1DE1E4D', '#cccccc')
```

## A011. Which position do you have in your institution?

```python
# Spalten mit Namen, die "A011" enthalten, anzeigen
df_A011 = df_cleanquest_2019.filter(like='A011')
df_A011
```

```python
values_A011=df_values_2019[df_values_2019["VAR"]=="A011"]

df_cleanquest_2019["A011"] = df_cleanquest_2019["A011"].astype(str)
values_A011["RESPONSE"] = values_A011["RESPONSE"].astype(str)

# Zahlen in den Zellen gegen die Bedeutung ersetzen
df_cleanquest_2019["A011-m"]=df_cleanquest_2019["A011"].map(values_A011.drop_duplicates("RESPONSE").set_index("RESPONSE")["MEANING"])
df_cleanquest_2019["A011-m"]
```

```python
# Zellen mit gleichem Inhalt zählen
value_counts = df_cleanquest_2019['A011-m'].value_counts()

# Neuen df erzeugen
df_A011 = df_cleanquest_2019['A011-m'].value_counts().to_frame()

# Spalten benennen
df_A011.reset_index(inplace=True)
df_A011.rename(columns={'index': 'position', 'A011-m': 'quantity'}, inplace=True)

# Freitextantworten und Häufigkeit zur Frage anzeigen ->Spalte A011_07
df_A011_a = df_cleanquest_2019['A011_07'].value_counts().to_frame()

# Spalten benennen
df_A011_a.reset_index(inplace=True)

df_A011_a.rename(columns={'index': 'position', 'A011_07': 'quantity'}, inplace=True)

print(df_A011_a)
```

```python
# Werte mit gleicher Bedeutung zusammenführen
df_A011_a['position'] = df_A011_a['position'].apply(lambda x: 'Professor' if 'prof' in x.lower() else 
                                                               ('Mid-level academic position' if 'group' in x.lower() or 'institut' in x.lower() or 'habil' in x.lower() else 
                                                                ('Researcher in Industry' if 'industry' in x.lower() else
                                                                 ('Technichian/ Support' if 'techn' in x.lower() or 'assistant' in x.lower() or 'staff' in x.lower() else
                                                                  ('Senior Scientist/ Director' if 'scien' in x.lower() or 'dire' in x.lower() else x)))))
# Groupby und Summierung durchführen
df_A011_grouped = df_A011_a.groupby('position')['quantity'].sum().reset_index()

# Ergebnisse nach Häufigkeit sortieren
df_A011_sorted = df_A011_grouped.sort_values(by='quantity', ascending=False)

print(df_A011_sorted)
```

```python
# DataFrames anhand des gemeinsamen Schlüssels 'position' zusammenführen
df_A011_merge = pd.merge(df_A011_sorted, df_A011, on='position', how='outer')

# Spalten quantity_1 und quantity_2 zusammenführen und Werte summieren
df_A011_merge['merged_quantity'] = df_A011_merge[['quantity_x', 'quantity_y']].sum(axis=1)

# doppelte Werten in der Spalte "merged_affiliation" entfernen
df_A011_merge.drop_duplicates(subset='position', inplace=True)

# ursprüngliche Spalten löschen
df_A011_merge.drop(['quantity_x', 'quantity_y'], axis=1, inplace=True)

df_A011_merge
```

```python
# Zeile mit Index 8 und 12 löschen, weil sie in den Freitextantworten aufgeschlüsselt wurde und als solche eingefügt wurden.
rows_to_delete = [8,12]

df_A011_merge = df_A011_merge.drop(rows_to_delete)
df_A011_merge
```

```python
# Zeilen mit einer Quantity von über 4 filtern
df_A011_merge_filtered = df_A011_merge[df_A011_merge['merged_quantity'] > 4]

# Spalte "Position" als Index setzen
df_A011_merge_filtered.set_index('position', inplace=True)

fig, ax = plt.subplots(figsize=(10, 6))  # Hier kannst du die Größe des Diagramms anpassen

# Daten als horizontales Balkendiagramm plotten
bars_A011 = df_A011_merge_filtered.plot.barh(ax=ax, color=petrol_rgb)

# absoluten Werte in die Balken einfügen
for bar in bars_A011.patches:
    ax.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, int(bar.get_width()),
            va='center', ha='left', color=orange_rgb)

plt.xlabel('Quantity')
plt.ylabel('Position')
plt.title('Positions of the participants')

# Diagramm anzeigen
plt.show()
```

## A12. To which discipline do you assign yourself?

```python
# Spalten mit Namen, die "A012" enthalten, anzeigen
df_A012 = df_cleanquest_2019.filter(like='A012')
df_A012
```

```python
# Leeres Dictionary für die Werte erstellen
values_dict = {}

# For-Schleife
for col_name in df_A012:
    values_dict[col_name] = df_values_2019[df_values_2019["VAR"] == col_name]

# Ergebnisse anzeigen
for col_name, values_df in values_dict.items():
    print(f"Values for {col_name}:")
    print(values_df)
    print()
```

```python
df_variables_2019 = pd.DataFrame (variables_2019)

# Zeilen von Index 12 bis 21 anzeigen
selected_rows_A012 = df_variables_2019.iloc[11:22]
selected_rows_A012
```

```python
columns_to_process = ['A012','A012_01','A012_02','A012_03','A012_04','A012_05','A012_06','A012_07','A012_08','A012_09','A012_09a']

for column in columns_to_process:
    values_column = df_values_2019[df_values_2019["VAR"] == column]
    
    df_cleanquest_2019[column] = df_cleanquest_2019[column].astype(str)
    values_column["RESPONSE"] = values_column["RESPONSE"].astype(str)
```

```python
# Spalten umbenennen, so dass Teildisziplin bekannt wird
# Dictionary erstellen
col_name_clear_A012 = {
    'A012': 'Ausweichoption (negativ) oder Anzahl',
    'A012_01': 'Inorganic chemistry',
    'A012_02': 'Organic chemistry',
    'A012_03': 'Physical chemistry',
    'A012_04': 'Chemical engineering',
    'A012_05': 'Theoretical chemistry',
    'A012_06': 'Pharmacy/Pharmaceutical chemistry',
    'A012_07': 'Material science',
    'A012_08': 'Biochemistry',
    'A012_09': 'Other',
    'A012_09a': 'open text'
}

df_A012 = df_A012.rename(columns=col_name_clear_A012)


# Anzahl der Vorkommen von 2 in den Spalten A012_01 - A012_09 zählen
df_A012_count = df_A012.iloc[:, 1:10].apply(lambda col: (col == '2').sum()).to_frame()

# Spalten umbenennen
df_A012_count.reset_index(inplace=True)
df_A012_count.rename(columns={'index': 'subdiscipline', 0: 'quantity'}, inplace=True)

# Zeilen mit quantity == 0 entfernen
df_A012_count = df_A012_count[df_A012_count['quantity'] != 0]

df_A012_count
```

```python
# Spalte "subdiscipline" als Index setzen
df_A012_count.set_index('subdiscipline', inplace=True)

df_A012_count.plot.barh(color=petrol_rgb)

plt.xlabel('Quantity')
plt.ylabel('Subdiscipline')
plt.title('Subdisciplines of chemistry')

# Diagramm anzeigen
plt.show()
```

```python
# Freitextantworten und Häufigkeit zur Frage anzeigen ->Spalte A004_13a
df_A012_a = df_cleanquest_2019['A012_09a'].value_counts().to_frame()

# Spaltennamen anpassen
df_A012_a.reset_index(inplace=True)

df_A012_a.rename(columns={'index': 'subdiscipline', 'A012_09a': 'quantity'}, inplace=True)
df_A012_a = df_A012_a.drop(0)

print(df_A012_a)
```

```python
# DataFrames anhand des gemeinsamen Schlüssels 'subdiscipline' zusammenführen
df_A012_merge = pd.merge(df_A012_count, df_A012_sorted, on='subdiscipline', how='outer')

# Spalten quantity_x und quantity_y zusammenführen und Werte summieren
df_A012_merge['merged_quantity'] = df_A012_merge[['quantity_x', 'quantity_y']].sum(axis=1)

# doppelte Werte in der Spalte "merged_affiliation" entfernen
df_A012_merge.drop_duplicates(subset='subdiscipline', inplace=True)

# ursprüngliche Spalten löschen
df_A012_merge.drop(['quantity_x', 'quantity_y'], axis=1, inplace=True)

df_A012_merge
```
